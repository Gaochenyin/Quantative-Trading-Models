```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
myfont = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 23,
}
```

```python
# Loading Pb Dominant
Dominant_Info = pd.read_csv('Dominant_Pb.csv')
Dominant_Info['start']
Dominant_Info['start'] = [datetime.strptime(it, '%Y-%m-%d') for it in Dominant_Info['start']]
Dominant_Info['end'] = [datetime.strptime(it, '%Y-%m-%d') for it in Dominant_Info['end']]
Dominant_Info['contract']= [contract.lower() for contract in Dominant_Info['contract']]
Dominant_Info
```

```python
def calcluate_dominant():
    files = os.listdir()
    dat = pd.DataFrame()
    for file in files:
        if file[-3:]=='csv' and file[:2]=='pb':
            print('loading '+file)
            temp_dat = pd.read_csv(file)
            try:
                dat = pd.concat([dat,temp_dat],axis=0)
            except:
                dat = temp_dat
    dat = dat.sort_values(by='平仓时间')
    dat['pnl']=dat['价差']*dat['开仓方向']
    dat['cumpnl'] = np.cumsum(dat['pnl'])
    dat
    dat = dat[['平仓时间','cumpnl','合约名称','pnl']]
    dat['合约名称'] = [name[:6] for name in dat['合约名称']]
    dat.drop_duplicates(['平仓时间'],inplace=True)
    dat.columns = ['Date','Cumnl','Name','pnl']
    dat['Date'] = [datetime.strptime(it, '%Y-%m-%d %H:%M:%S') for it in dat['Date']]
    M = len(dat)
    dat.index = range(M)
    for i in dat.index:
        temp_dat = dat.loc[i]
        temp_name = temp_dat['Name']
        try:
            dominant_name = Dominant_Info.iloc[np.argwhere(Dominant_Info['contract']==temp_name)[0],:]
            if temp_dat['Date']>dominant_name['end'].values[0] or temp_dat['Date']<dominant_name['start'].values[0]:
                dat.drop(i,inplace=True)
        except IndexError:
            dat.drop(i,inplace=True)
    dat.plot(x='Date',y='Cumnl',figsize=(20,10))
    plt.tick_params(labelsize=15)
    plt.xlabel('Time',myfont)
    plt.ylabel('Cumulative Return',myfont)
    plt.legend(loc='upper left',fontsize=15)
    ret = np.sum(dat['pnl'])
    sharp = np.mean(dat['pnl'])/np.std(dat['pnl'])
    return dat,(ret,sharp)
```

```python
dat,info = calcluate_dominant()
print('Cumulative Return is {0:0.3f}, Sharp Ratio is {1:0.3f}'.format(info[0],info[1]))
```
